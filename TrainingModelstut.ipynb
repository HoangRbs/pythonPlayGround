```python
# since we already have Id as index, we don't need nother index that was generated by pandas
data_df = pd.read_csv('./housingprice_train.csv', index_col='Id')
data_df
```

```python
data_df.columns
```

```python
chosen_features = [ # Những feature tiêu biểu ảnh hưởng mạnh đến dự doán giá nhà
    'LotArea', # diện tích mảnh đất, càng lớn càng đắt 
    'YearBuilt', # năm xây nhà, nhà mới xây thì thường có giá cao hơn
    '1stFlrSF', # diện tích lầu 1 là bao nhiêu
    '2ndFlrSF',
    'FullBath', # số phòng tắm
    'BedroomAbvGr', # phòng ngủ đạt trên tiêu chuẩn
    'TotRmsAbvGrd', # tổng số phòng trên tiêu chuẩn
]
```

```python
X = data_df[chosen_features]
X.head()
```

```python
y = data_df['SalePrice']
y.head()
```

```python
type(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
# Training machine learning model
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeRegressor(random_state=1)
dt_model.fit(X_train, y_train)
```

```python
y_preds = dt_model.predict(X_valid.head())

pd.DataFrame({
    'y': y_valid.head(),
    'y_preds': y_preds
})
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train, y_train)
```

```python
y_preds = rf_model.predict(X_valid.head())
```

```python
pd.DataFrame({
    'actual values': y_valid.head(),
    'predicted values': y_preds
})
```

```python
# predict a new one input
X_valid.head()
```

```python
rf_model.predict([
    [6969, 2022, 1000, 800, 4, 5, 8]
])
```

```python
# model evaluation (Regression)

rf_model.score(X_valid, y_valid)
```

```python
# mean absolute error
from sklearn.metrics import mean_absolute_error

y_preds = rf_model.predict(X_valid)
```

```python
result_df = pd.DataFrame({
    'actual values': y_valid,
    'predicted values': y_preds
})

result_df
```

```python
mae = mean_absolute_error(y_valid, y_preds)
```

```python
mae

# mean absolute error is: 
# SUM of all rows 'difference' column: absolute (actualval - predicted) 
# then divide by all rows (which is mean)
# --> it measure HOW FAR the predicted value to the actual value (by average)
```
