```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./california_cities.csv')
cities.head()
```

```python
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

plt.style.use('seaborn')
plt.figure(figsize=(8, 6))
plt.scatter(lon, lat,
            c=np.log10(population),
            cmap='viridis',
            s=area,
            alpha=0.5
            )
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)');
plt.clim(3, 7); # limit color

plt.title('califonia cities: population and area distribution')

# create legends (city area annotation) for cities size
area_range = [50, 100, 300, 500] # km^2

for val in area_range:
    plt.scatter([], [], # we don't want any point to show on the map since we just want annotation - with label of course
                s=val, # we're scattering here, so the point (not display) are type of circles ... with SIZE, --> so are the annotations
                label=str(val) + 'km$^2$',
                c='k',
                alpha=0.4
                )
    
plt.legend(labelspacing=1) # show annotations
```
