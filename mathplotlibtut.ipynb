```python
x = np.linspace(0, 10, 1000) # 1000 inputs from 0 to 10 -- with equally distances
plt.plot(x, np.sin(x), linestyle='dashed', label = 'sin(x)') 
plt.plot(x, np.cos(x), color = "red", label = 'cos(x)')

plt.title('sin(x) and cos(x)')
plt.xlabel('bien x')
plt.ylabel('F(x)')
# plt.xlim([0, 4]) # limit x
# plt.axis([xmin, xmax, ymin, ymax])
plt.axis('equal')
# plt.axis('tight')

plt.legend(); # show labels annotation of each line
```

```python
x = [1,2,3,4]
y = [11,22,33,44]

fig, ax = plt.subplots(figsize=(4,5)) # figure size = width & height of the plot
# plot the data
ax.plot(x, y)
ax.set(
    title='A simple plot',
    xlabel='x-axis',
    ylabel='y-axis'
);
```

```python
rng = np.random.RandomState(42)

x = rng.randn(100)
y = rng.randn(100)
sizes = rng.rand(100) * 1000 # size for each point (x, y)
colors = rng.rand(100) # color for each point

fig,ax = plt.subplots()
img1 = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.3)

fig.colorbar(img1) # the figure need to know which graph in it to show colors bar
```

```python
# prepare student height data

mean_height = 170 # chieu cao trung binh 1m7
std_height = 10 # do lech chuan ve chieu cao: dao ng trn di 10
total_students = 250

student_heights = np.random.normal(mean_height, std_height, total_students)
```

```python
student_heights[:10]
```

```python
# show histogram
plt.hist(student_heights, bins=30);
```

```python
# subplots

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows=2, ncols=2, # 2 rows, 2 cols --> 4 graphs
    figsize=(10, 5) # each graph, width = 10, height = 5
);

# plot data to each axis
ax1.plot(x, x/2)
ax1.plot(x, x*2)

ax2.scatter(np.random.random(10), np.random.random(10));

ax3.bar(drink_prices.keys(), drink_prices.values());

ax4.hist(student_heights, bins=30);
```

```python
# subplots

fig, ax = plt.subplots(
    nrows=2, 
    ncols=2, # 2 rows, 2 cols --> 4 graphs
    figsize=(10, 5) # each graph, width = 10, height = 5
);

# plot data to each axis
ax[0,0].plot(x, x/2)
ax[0,0].plot(x, x*2)

ax[0,1].scatter(np.random.random(10), np.random.random(10));

ax[1,0].bar(drink_prices.keys(), drink_prices.values());

ax[1,1].hist(student_heights, bins=30);
```
